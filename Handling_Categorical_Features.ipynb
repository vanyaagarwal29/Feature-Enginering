**<h2>One hot encoding with many categories in a feature</h2>**

```python
df=pd.read_csv('mercedes.csv',usecols=["X0",'X1','X2','X3','X4','X5','X6'])
df.head()
```

```python
for i in df.columns:
  print(len(df[i].unique()))
```

```python
#We will take 10 most frequent occuring variable of that feature
df.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
lst_10=df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
lst_10
```

```python
import numpy as np
for categories in lst_10:
  df[categories]=np.where(df['X1']==categories,1,0)

lst_10.append('X1')
df[lst_10]
```

**<h2>Count or Frequency Encoding</h2>**

```python
train_set = pd.read_csv('adult.data', header = None,index_col=None) 
train_set.head()
```

```python
columns=[1,3,5,6,7,8,9,13]
train_set=train_set[columns]
```

```python
train_set.columns=['Employment','Degree','Status','Designation','family-job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]:
  print(feature,":",len(train_set[feature].unique()),'labels')
```

```python
country_map=train_set['Country'].value_counts().to_dict()
country_map
```

```python
train_set['Country']=train_set['Country'].map(country_map)
train_set.head()
```

Advantages
1. Easy To Use
2.Not increasing feature space

Disadvantages
1. It will provide same weight if the frequencies are same

**<h2>Target Guided Ordinal Encoding</h2>**
- Ordering the labels according to the target
- Replace the labels by the joint probability of being 1 or 0

```python
df=pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
df.head()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head()
```

**<h2>Mean Encoding</h2>**

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
mean_ordinal
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
df.head()
```

Advantages
1. It captures information within the label therefore lendering more predictive features.
2. It creates a monotonic relationship between variable and target.

Disadvantage
1. It leads to overfitting

**<h2>Probability Ratio Encoding</h2>**

```python
df=pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
df.head()
```

```python

df['Cabin'].fillna('Missing', inplace=True)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df.Cabin.unique()
```

```python
prob_df=df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died']=1-prob_df['Survived']
prob_df.head()
```

```python
prob_df['Probability_ratio']=prob_df['Survived']/prob_df['Died']
prob_df.head()
```

```python
probability_encoding=prob_df['Probability_ratio'].to_dict()
```

```python
df['Cabin_encoded']=df['Cabin'].map(probability_encoding)
df.head()
```
