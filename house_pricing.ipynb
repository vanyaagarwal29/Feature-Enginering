```python
features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1]
for feature in features_nan:
    print("The faeture is {} and missing values in {}%".format(feature,np.round(dataset[feature].isnull().mean()*100,4)))
```

```python
##Replacing categorical features
categorical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
categorical_feature
```

```python
data=dataset.copy()
def replace_nan_features(dataset,categorical_feature):
    data[categorical_feature]=data[categorical_feature].fillna('Missing')
    return data
```

```python
replace_nan_features(dataset,categorical_feature)
```

```python
data=dataset.copy()
def replace_nan_features(dataset,categorical_feature):
    data[categorical_feature]=data[categorical_feature].fillna(data[categorical_feature].mode())
    return data
```

```python
replace_nan_features(dataset,categorical_feature)
```

here we can see that we still have nan values so we will not use this approach

```python
numerical_features_nan=[feature for feature in dataset.columns if dataset[feature].dtype!='object' and dataset[feature].isnull().sum()>=1]
```

```python
numerical_features_nan
```

```python
for feature in numerical_features_nan:
    #since thera are outliers we are going to replae with median
    median_values=dataset[feature].median()
    dataset[feature+'_nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_values,inplace=True)
```

```python
dataset
```

```python
#temporal variable
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature+'_age of house']=dataset['YrSold']-dataset[feature]
```
